```python
import pandas as pd
import numpy as np
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("student_dataset.csv")
print("Original Dataset:")
print(df)

# Handle Missing Data
df.replace("*Missing*", np.nan, inplace=True)
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['City'] = df['City'].fillna("Unknown")
print("\nAfter Handling Missing Data:")
print(df)

# Handle Noisy Data
df.loc[df['Marks'] < 0, 'Marks'] = df['Marks'].median()
df.loc[df['Marks'] > 100, 'Marks'] = 100
print("\nAfter Handling Noisy Data:")
print(df)

# Data Transformation
df['Income_Normalized'] = (df['Income'] - df['Income'].min()) / (df['Income'].max() - df['Income'].min())
print("\nAfter Normalization:")
print(df)

# Encoding Categorical Data
df = pd.get_dummies(df, columns=['Gender', 'City'], drop_first=True)

print("\nAfter Encoding Categorical Data:")
print(df)

# Save Preprocessed Dataset
df.to_csv("student_dataset_cleaned.csv", index=False)
print("\n✅ Preprocessed dataset saved as 'student_dataset_cleaned.csv'")
files.download("student_dataset_cleaned.csv")
```

# Task
Show an example of training a simple ML model.

## Generate synthetic data

### Subtask:
Create a simple dataset for training and testing.


**Reasoning**:
The instructions require creating a synthetic dataset and splitting it into training and testing sets. This involves generating random numbers, applying a linear relationship with noise, and then manually splitting the resulting arrays. All these steps can be accomplished in a single code block using numpy.



```python
import numpy as np

# Generate 1000 random floating-point numbers between 0 and 10 for the feature x
x = 10 * np.random.rand(1000)

# Generate the target variable y with noise
noise = np.random.randn(1000)
y = 2 * x + 1 + noise

# Split the data into training and testing sets (80% train, 20% test)
split_index = int(0.8 * len(x))

x_train = x[:split_index]
y_train = y[:split_index]
x_test = x[split_index:]
y_test = y[split_index:]

print(f"Shape of x_train: {x_train.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of x_test: {x_test.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

## Define and compile the model

### Subtask:
Build a simple neural network model using TensorFlow/Keras and compile it.


**Reasoning**:
Build a simple neural network model using TensorFlow/Keras and compile it.



```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

# Create a Sequential model instance
model = Sequential()

# Add a Dense layer with 1 neuron and specify input_shape
model.add(Dense(1, input_shape=(1,)))

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Print a summary of the model
model.summary()
```

## Train the model

### Subtask:
Train the defined model using the generated data.


**Reasoning**:
Train the model using the training data and store the training history.



```python
history = model.fit(x_train, y_train, epochs=50)
```

## Evaluate the model

### Subtask:
Evaluate the trained model on the test data to assess its performance.


**Reasoning**:
Evaluate the trained model on the test data to assess its performance and print the result.



```python
loss = model.evaluate(x_test, y_test)
print(f"Test Loss: {loss}")
```
